```python
from abc import ABC
from collections import namedtuple
import os
```

```python
class Tag:
    
    def __init__(self, word, start, length, tag):
        self.word = word
        self.start = start
        self.length = length
        self.tag = tag
```

```python
def load_labels(data):
    lines = None
    with open('labels.txt') as labels:
        lines = labels.readlines()
    labels = []
    for line, seq in zip(lines, data):
        split = iter(line.split())
        tags = []
        while True:
            token = next(split)
            if token == 'EOL':
                break
            else:
                start = int(token)
                length = int(next(split))
                tag = next(split)
                word = seq[start:start + length]
                tags.append(Tag(start=start, length=length, tag=tag, word=word))
        labels.append(tags)
    return labels
```

```python
def load_data1():
    mapping = dict()
    for file in os.listdir('data1'):
        if file.endswith('ann'):
            with open(os.path.join('data1', file)) as f:
                for line in f.readlines():
                    split = line.strip().split('\t')
                    tag = split[1][:3]
                    if tag == 'PER':
                        tag = 'PERSON'
                    if tag == 'ORG' or tag == 'PERSON':
                        for w in split[2].split():
                            mapping[w.lower()] = tag
    return mapping
```

```python
def load_data(file):
    with open(file) as f:
        return list(map(lambda x: x.lower(), f.readlines()))
    
```

```python
separators = ['.', ',', '—', ' ', '\n', '?', '!', '"', '(', ')']

def parse_words(line):
    words = []
    word = []
    for i in range(len(line)):
        if line[i] not in separators:
            word.append(line[i])
        else:
            if word:
                words.append(Tag(word=''.join(word), tag=None, start=i - len(word), length=len(word)))
                word = []
    return words
```
